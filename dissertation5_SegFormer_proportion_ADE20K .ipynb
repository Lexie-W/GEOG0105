```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model.to(device)
model.eval()

label_ids_to_calculate = [
    1,   # wall
    2,   # building
    4,   # tree
    6,   # road
    11,  # sidewalk
    12,  # person;individual;someone;somebody;mortal;soul
    13,  # ground
    20,  # car
    21,  # water
    32,  # fence
    43,  # signboard
    48,   # skyscraper
    52,  # path
    53,  # stairs
    59,  # staircase
    61,  # bridge
    69,  # bench
    72,  # palm tree
    86,  # awning
    87,  # streetlight
    93,  # pole
    94,  # land/soil
    95,  # handrail
    96,   # escalator
    121,  # step;stair
    136,  # traffic light
    138,  # trash can; ashcan
    127,  # bicycle
    114,  # tent;collapsible shelter
    102,  # van
    106,  # canopy
]
output_dir = "/content/drive/MyDrive/GEOG0105/Output"
os.makedirs(output_dir, exist_ok=True)
```

```python
@torch.no_grad()
def calculate_all_proportions(image, label_ids):
    inputs = image_processor(images=image, return_tensors="pt").to(device)
    outputs = model(**inputs)
    logits = outputs.logits
    predicted = torch.argmax(logits, dim=1).squeeze()

    total = predicted.numel()
    return [(predicted == lid).sum().item() / total for lid in label_ids]
```

```python
from tqdm import tqdm
new_rows = []

# sample_images assumed to be already loaded
for idx, row in tqdm(sample_images.iloc[:12800].iterrows(), total=12800):
    filename = row['filename_x']
    image_path = os.path.join("/content/drive/MyDrive/GEOG0105/GSV_final", filename)

    if not os.path.exists(image_path):
        print(f"Missing file: {filename}, skipping.")
        continue

    try:
        with Image.open(image_path).convert("RGB") as img:
            proportions = calculate_all_proportions(img, label_ids_to_calculate)

        row_data = row.to_dict()
        for i, lid in enumerate(label_ids_to_calculate):
            row_data[str(lid)] = proportions[i]
        new_rows.append(row_data)

        # Save every 1000 images
        if len(new_rows) % 1000 == 0:
            temp_df = pd.DataFrame(new_rows)
            temp_df.to_csv(f"{output_dir}/intermediate_{idx}.csv", index=False)
            print(f"Saved intermediate result at image {idx}")

    except Exception as e:
        print(f"Error processing {filename}: {e}")
        continue
```

```python
def process_image(image_path):
    try:
        image = Image.open(image_path).convert("RGB")
    except Exception as e:
        print(f"Error reading image {image_path}: {e}")
        return None

    inputs = image_processor(images=image, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits  # (1, num_classes, H, W)

    upsampled_logits = torch.nn.functional.interpolate(
        logits,
        size=image.size[::-1],  # (height, width)
        mode="bilinear",
        align_corners=False
    )

    predicted = upsampled_logits.argmax(dim=1)[0].cpu().numpy()

    total_pixels = predicted.size
    proportions = {
        str(label): (predicted == label).sum() / total_pixels
        for label in label_ids_to_calculate
    }

    return proportions
```

```python
new_results = []

for _, row in failed_rows.iterrows():
    img_path = os.path.join(save_dir, row["filename_x"])
    proportions = process_image(img_path)
    if proportions:
        result_row = row.to_dict()
        result_row.update(proportions)
        new_results.append(result_row)

new_df = pd.DataFrame(new_results)
new_df
```
